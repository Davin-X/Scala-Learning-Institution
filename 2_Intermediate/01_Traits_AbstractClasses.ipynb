## ðŸ›ï¸ Abstract Classes

Abstract classes define incomplete blueprints that subclasses must complete.

```scala
// Abstract class definition
abstract class Animal(val name: String) {
  // Abstract methods (no implementation)
  def speak(): Unit
  def move(): Unit
  
  // Concrete method
  def describe(): String = s"This is $name"
}

// Concrete subclass
class Dog(name: String) extends Animal(name) {
  def speak(): Unit = println("Woof!")
  def move(): Unit = println("Runs on four legs")
  
  // Override concrete methods
  override def describe(): String = s"Dog named $name"
}

class Bird(name: String) extends Animal(name) {
  def speak(): Unit = println("Tweet!")
  def move(): Unit = println("Flies through air")
}

// Usage
val animals = List(new Dog("Buddy"), new Bird("Tweety"))
animals.foreach { animal =>
  println(animal.describe())
  animal.speak()
  animal.move()
  println()
}
```

## ðŸ§© Traits

Traits are Scala's flexible units of behavior that can be mixed into classes.

```scala
// Basic traits
trait Speaker {
  def speak(message: String): Unit
  
  // Default implementation
  def greet(): Unit = speak("Hello!")
}

trait Mover {
  def move(distance: Double): Unit
}

trait Logger {
  def log(message: String): Unit = println(s"[LOG] $message")
}

// Class using multiple traits
class Robot extends Speaker with Mover with Logger {
  def speak(message: String): Unit = println(s"Robot: $message")
  
  def move(distance: Double): Unit = {
    log(f"Moving $distance%.1f units")
    println(f"Robot moved $distance%.1f units")
  }
  
  // Override trait methods
  override def log(message: String): Unit =
    println(s"[ROBOT LOG] $message")
}

val robot = new Robot()
robot.greet()
robot.move(10.5)
robot.log("Task complete")
```

## ðŸ”— Trait Composition

Traits can be mixed into classes at instantiation time and stacked together.

```scala
// Stackable traits
abstract class Beverage {
  val description: String = "Beverage"
  val cost: Double = 0.0
  
  def getDescription: String = description
  def getCost: Double = cost
}

trait MilkAdder extends Beverage {
  abstract override def getDescription: String =
    super.getDescription + " with milk"
  abstract override def getCost: Double =
    super.getCost + 0.50
}

trait SugarAdder extends Beverage {
  abstract override def getDescription: String =
    super.getDescription + " with sugar"
  abstract override def getCost: Double =
    super.getCost + 0.25
}

// Stackable behavior at instantiation
class Coffee extends Beverage {
  override val description = "Coffee"
  override val cost = 2.00
}

val plainCoffee = new Coffee()
val coffeeWithMilk = new Coffee with MilkAdder
val fancyCoffee = new Coffee with MilkAdder with SugarAdder

println("Coffee Variations:")
println(f"${plainCoffee.getDescription}%-20s: $${plainCoffee.getCost}%.2f")
println(f"${coffeeWithMilk.getDescription}%-20s: $${coffeeWithMilk.getCost}%.2f")
println(f"${fancyCoffee.getDescription}%-20s: $${fancyCoffee.getCost}%.2f")
```

## ðŸ›ï¸ Design Patterns with Traits

Traits enable powerful design patterns for modular architecture.

```scala
// Strategy Pattern
trait TaxStrategy {
  def calculateTax(amount: Double): Double
}

class StandardTax extends TaxStrategy {
  def calculateTax(amount: Double): Double = amount * 0.1  // 10%
}

class LuxuryTax extends TaxStrategy {
  def calculateTax(amount: Double): Double = amount * 0.25  // 25%
}

class Order(val amount: Double) {
  var taxStrategy: TaxStrategy = new StandardTax()
  
  def setTaxStrategy(strategy: TaxStrategy): Unit = {
    taxStrategy = strategy
  }
  
  def totalCost(): Double = {
    amount + taxStrategy.calculateTax(amount)
  }
}

val standardOrder = new Order(100.0)
val luxuryOrder = new Order(100.0)
luxuryOrder.setTaxStrategy(new LuxuryTax())

println(f"Standard order: $$${standardOrder.totalCost()}%.2f")
println(f"Luxury order: $$${luxuryOrder.totalCost()}%.2f")
```

## ðŸ’Ž Self Types

Self types ensure that a trait can only be mixed into classes that provide required capabilities.

```scala
// Self types for dependency injection
trait UserRepository {
  def findUser(id: Long): Option[String]
}

trait UserService {
  // Self type: must be mixed into something that has UserRepository
  this: UserRepository =>
  
  def authenticateUser(id: Long): Boolean = {
    findUser(id) match {
      case Some(name) => 
        println(s"Authenticated: $name")
        true
      case None => 
        println("User not found")
        false
    }
  }
}

// Concrete implementation
class UserManagement extends UserRepository with UserService {
  private val users = Map(1L -> "Alice", 2L -> "Bob")
  
  def findUser(id: Long): Option[String] = users.get(id)
}

val userService = new UserManagement()
userService.authenticateUser(1L)
userService.authenticateUser(999L)
```

Create a shape hierarchy using abstract classes and traits.

```scala
// Exercise 1: Shape System
// FIXME: Replace ??? with your code

// Abstract shape class
abstract class Shape {
  def area(): Double
  def perimeter(): Double
}

// Color trait
trait Colored {
  val color: String
  def describeColor(): String = s" colored $color"
}

// Movable trait
trait Movable {
  var x: Double = 0
  var y: Double = 0
  
  def move(dx: Double, dy: Double): Unit = {
    x += dx
    y += dy
  }
  
  def position(): String = f"($x%.1f, $y%.1f)"
}

// Shape implementations
class Circle(radius: Double) extends Shape with Colored with Movable {
  val color = "red"
  
  def area(): Double = math.Pi * radius * radius
  def perimeter(): Double = 2 * math.Pi * radius
  
  def describe(): String =
    f"Circle with radius $radius%.1f at ${position()}${describeColor()}, area=${area()}%.2f, perimeter=${perimeter()}%.2f"
}

class Rectangle(width: Double, height: Double) extends Shape with Colored with Movable {
  val color = "blue"
  
  def area(): Double = width * height
  def perimeter(): Double = 2 * (width + height)
  
  def describe(): String =
    f"Rectangle ${width}%.1fx${height}%.1f at ${position()}${describeColor()}, area=${area()}%.2f, perimeter=${perimeter()}%.2f"
}

// Test your shape system
println("Shape System Exercise:")
println("=" * 25)

val circle = new Circle(5.0)
val rectangle = new Rectangle(4.0, 6.0)

circle.move(10, 20)
rectangle.move(5, 15)

val shapes = List(circle, rectangle)
shapes.foreach(shape => println(shape.describe()))
```

### Exercise 2: Notification System

Build a notification system using traits and stackable behavior.

```scala
// Exercise 2: Notification System
// FIXME: Replace ??? with your code

// Base notification
abstract class Notification {
  val recipient: String
  val message: String
  
  def send(): Unit
}

// Delivery traits
trait EmailDelivery extends Notification {
  def send(): Unit = {
    super.send()  // Call next in stack
    println(s"ðŸ“§ Email to $recipient: $message")
  }
}

trait SMSDelivery extends Notification {
  def send(): Unit = {
    super.send()
    println(s"ðŸ“± SMS to $recipient: ${message.take(160)}")
  }
}

trait PushDelivery extends Notification {
  def send(): Unit = {
    super.send()
    println(s"ðŸ”” Push notification to $recipient: $message")
  }
}

// Alert levels
trait UrgentModifier extends Notification {
  abstract override def send(): Unit = {
    super.send()
    println(s"ðŸš¨ URGENT: Follow-up required for $recipient")
  }
}

trait ScheduledModifier extends Notification {
  val delayMinutes: Int
  abstract override def send(): Unit = {
    println(s"â° Scheduled: Will send in $delayMinutes minutes to $recipient")
    // In real system, this would schedule
  }
}

// Concrete notification types
class BasicNotification(recipient: String, message: String)
  extends Notification
  with EmailDelivery

class FullNotification(recipient: String, message: String)
  extends Notification
  with EmailDelivery
  with SMSDelivery
  with PushDelivery

class UrgentNotification(recipient: String, message: String)
  extends Notification
  with EmailDelivery
  with SMSDelivery
  with UrgentModifier

// Test the notification system
println("Notification System Exercise:")
println("=" * 32)

println("\nBasic notification:")
val basic = new BasicNotification("user@example.com", "Welcome!")
basic.send()

println("\nFull notification:")
val full = new FullNotification("customer@company.com", "Order processed")
full.send()

println("\nUrgent notification:")
val urgent = new UrgentNotification("admin@alerts.com", "Server down")
urgent.send()

println("\nAd-hoc composition:")
val customNotification = new BasicNotification("test@test.com", "Test message")
  with SMSDelivery
customNotification.send()
```

### Exercise 3: Cake Pattern Component System

Implement dependency injection using self-types and traits.

```scala
// Exercise 3: Cake Pattern
// FIXME: Replace ??? with your code

// Component traits
trait DatabaseComponent {
  val database: Database
  trait Database {
    def save(item: String): Unit
    def findAll(): List[String]
  }
}

trait LoggerComponent {
  val logger: Logger
  trait Logger {
    def log(message: String): Unit
  }
}

trait CacheComponent {
  val cache: Cache
  trait Cache {
    def set(key: String, value: String): Unit
    def get(key: String): Option[String]
  }
}

// Application services
trait DataService {
  this: DatabaseComponent with LoggerComponent with CacheComponent =>
  
  def saveItem(item: String): Unit = {
    logger.log(s"Saving item: $item")
    database.save(item)
    cache.set(item, item)
    logger.log(s"Item saved: $item")
  }
  
  def getAllItems(): List[String] = {
    logger.log("Retrieving all items")
    val items = database.findAll()
    logger.log(s"Found ${items.size} items")
    items
  }

  def getFromCache(key: String): Option[String] = {
    cache.get(key)
  }
}

// Implementations
trait InMemoryDatabase extends DatabaseComponent {
  val database = new Database {
    private var data = List[String]()
    def save(item: String): Unit = { data = item :: data }
    def findAll(): List[String] = data
  }
}

trait ConsoleLogger extends LoggerComponent {
  val logger = new Logger {
    def log(message: String): Unit = println(s"[APP] $message")
  }
}

trait SimpleCache extends CacheComponent {
  val cache = new Cache {
    private var storage = Map[String, String]()
    def set(key: String, value: String): Unit = {
      storage = storage + (key -> value)
    }
    def get(key: String): Option[String] = storage.get(key)
  }
}

// Application composition
class MyApplication extends DataService
  with InMemoryDatabase
  with ConsoleLogger
  with SimpleCache

// Test the cake pattern
println("Cake Pattern Dependency Injection:")
println("=" * 35)

val app = new MyApplication()

app.saveItem("item1")
app.saveItem("item2")
app.saveItem("item3")

println("\nAll items:")
app.getAllItems().foreach(println)

println("\nCache testing:")
println(s"Get from cache 'item1': ${app.getFromCache("item1")}")
println(s"Get from cache 'nonexistent': ${app.getFromCache("nonexistent")}")

println("\n==> Components are loosely coupled and easily testable!")
```

## ðŸ—ï¸ Function Basics

Functions are reusable blocks of code that perform specific tasks.

```scala
// Basic function definition
def greet(name: String): String = {
  s"Hello, $name! Welcome to Scala."
}

// Function call
val message1 = greet("Alice")
val message2 = greet("Bob")

println(message1)
println(message2)
```

## ðŸ” Higher-Order Functions

Functions that take other functions as parameters or return functions.

```scala
// Function that takes a function as parameter
def applyOperation(a: Int, b: Int, operation: (Int, Int) => Int): Int = {
  operation(a, b)
}

val add = (x: Int, y: Int) => x + y
val multiply = (x: Int, y: Int) => x * y

println(s"5 + 3 = ${applyOperation(5, 3, add)}")
println(s"5 * 3 = ${applyOperation(5, 3, multiply)}")

// Working with collections
val numbers = List(1, 2, 3, 4, 5)
val doubled = numbers.map(_ * 2)
val evens = numbers.filter(_ % 2 == 0)

println(s"Numbers: $numbers")
println(s"Doubled: $doubled")
println(s"Evens: $evens")
```

## ðŸ”„ Recursion

Functions that call themselves to solve problems.

```scala
// Factorial - classic recursion example
def factorial(n: Int): Int = {
  if (n <= 1) 1
  else n * factorial(n - 1)
}

// Test factorial
for (i <- 1 to 10) {
  println(f"$i! = ${factorial(i)}")
}

// Fibonacci
def fibonacci(n: Int): Int = {
  if (n <= 1) n
  else fibonacci(n - 1) + fibonacci(n - 2)
}

println("Fibonacci: " + (0 to 10).map(fibonacci).mkString(", "))
```

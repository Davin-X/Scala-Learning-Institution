## ðŸ’Ž **Semigroup: Combining Values**

**The simplest type class - knowing how to combine values of the same type**

```scala
// Semigroup: The foundation of all combination laws
import cats.Semigroup
import cats.implicits._

println("ðŸ’Ž SEMIGROUP: COMBINING VALUES WITH ASSURANCE")
println()

// Basic semigroup examples
val num1 = 5
val num2 = 3
val combined = Semigroup[Int].combine(num1, num2) // 8
println(s"Int semigroup: $num1 |+| $num2 = $combined")

val str1 = "Hello "
val str2 = "World!"
val greeting = str1 |+| str2
println(s"String semigroup: '$str1' |+| '$str2' = '$greeting'")

// List concatenation
val list1 = List(1, 2, 3)
val list2 = List(4, 5, 6)
val joined = list1 |+| list2
println(s"List semigroup: $list1 |+| $list2 = $joined")

// Custom semigroup - maximum operation
implicit val maxSemigroup: Semigroup[Int] = new Semigroup[Int] {
  def combine(a: Int, b: Int): Int = math.max(a, b)
}

val max1 = 10
val max2 = 25
val maximum = max1 |+| max2
println(s"Custom max semigroup: $max1 |+| $max2 = $maximum")

// Combining collections of semigroups
val numbers = List(1, 2, 3, 4, 5)
val sum = numbers.combineAll // Uses int semigroup (addition)
println(s"Numbers sum: ${numbers.mkString("+")} = $sum")

val strings = List("Functional", " ", "Programming", " ", "Rocks!")
val sentence = strings.combineAll // Uses string semigroup (concatenation)
println(s"String combination: $sentence")

println("\nâœ… Semigroup Properties (Laws):")
println("â€¢ Associativity: (a |+| b) |+| c == a |+| (b |+| c)")
println("â€¢ Closure: combine(a,b) returns same type as a and b")
println("â€¢ No identity element requirement (unlike Monoid)")
```

## ðŸŽ­ **Monoid: Semigroup + Identity**

**Semigroup with identity - the complete algebraic structure**

```scala
// Monoid: Semigroup + Identity Element
import cats.Monoid

println("ðŸŽ­ MONOID: SEMIGROUP WITH IDENTITY - COMPLETE ABSTRACTION")
println()

// Standard monoids
println("Standard Monoids:")
println(s"Int addition: empty = ${Monoid[Int].empty}, combine = ${(a: Int, b: Int) => a + b}")
println(s"Int multiplication: empty = ${Monoid[Int].empty(implicitly[Monoid[Int]])}, wait...") // This won't work as expected

// Custom monoid for different operations
case class Sum(value: Int)
case class Product(value: Int)

implicit val sumMonoid: Monoid[Sum] = new Monoid[Sum] {
  def empty: Sum = Sum(0)
  def combine(a: Sum, b: Sum): Sum = Sum(a.value + b.value)
}

implicit val productMonoid: Monoid[Product] = new Monoid[Product] {
  def empty: Product = Product(1)
  def combine(a: Product, b: Product): Product = Product(a.value * b.value)
}

val sums = List(Sum(1), Sum(2), Sum(3), Sum(4))
val total = sums.combineAll
println(s"Sum monoid: ${sums.map(_.value).mkString("+")} = ${total.value}")

val products = List(Product(2), Product(3), Product(4))
val result = products.combineAll
println(s"Product monoid: ${products.map(_.value).mkString("Ã—")} = ${result.value}")

// Option monoid - uses semigroup for inner type
val optionInts = List(Some(1), Some(2), None, Some(3))
val optionCombined = optionInts.combineAll
println(s"Option monoid: $optionInts -> $optionCombined")

// Map monoid - combines values at same keys
val map1 = Map("apples" -> 3, "oranges" -> 2)
val map2 = Map("apples" -> 5, "bananas" -> 1)
val combinedMap = map1 |+| map2
println(s"Map monoid: $map1 |+| $map2 = $combinedMap")

println("\nâœ… Monoid Laws:")
println("â€¢ Left identity: Monoid[A].empty |+| a == a")
println("â€¢ Right identity: a |+| Monoid[A].empty == a")
println("â€¢ Associativity: (a |+| b) |+| c == a |+| (b |+| c)")
println("â€¢ Closure: combine(a,b) returns same type")

println("\nðŸŽ¯ Monoid vs Semigroup:")
println("â€¢ Semigroup: Just combination (a |+| b)")
println("â€¢ Monoid: Combination + identity element")
println("â€¢ Use Monoid when you need an 'empty' or 'zero' value")
```

## ðŸŽ¨ **Functor: Context Transformation**

**Map over any context - the universal API transformation**

```scala
// Functor: Mapping over contexts
import cats.Functor

println("ðŸŽ¨ FUNCTOR: MAPPING OVER CONTEXTS - UNIVERSAL TRANSFORMATION")
println()

// Standard functor examples
println("Standard Functors:")

// List functor
val numbers = List(1, 2, 3, 4, 5)
val doubled = Functor[List].map(numbers)(_ * 2)
val squared = numbers.map(math.pow(_ * 1.0, 2).toInt) // Using built-in map
println(s"List functor: $numbers â†’ $doubled")
println(s"List map squared: $squared")

// Option functor
val maybeNumber: Option[Int] = Some(42)
val maybeString = Functor[Option].map(maybeNumber)(_.toString + " is the answer")
val noneResult = Functor[Option].map(None: Option[Int])(_.toString)
println(s"Option functor: $maybeNumber â†’ $maybeString")
println(s"Option map None: $noneResult")

// Either functor
val success: Either[String, Int] = Right(100)
val error: Either[String, Int] = Left("Something went wrong")
val successMapped = Functor[Either[String, *]].map(success)(_ + 50)
val errorMapped = Functor[Either[String, *]].map(error)(_.toString)
println(s"Either functor success: $success â†’ $successMapped")
println(s"Either functor error: $error â†’ $errorMapped")

// Custom functor for trees
sealed trait Tree[+A]
case class Leaf[A](value: A) extends Tree[A]
case class Branch[A](left: Tree[A], right: Tree[A]) extends Tree[A]

implicit val treeFunctor: Functor[Tree] = new Functor[Tree] {
  def map[A, B](tree: Tree[A])(f: A => B): Tree[B] = tree match {
    case Leaf(value) => Leaf(f(value))
    case Branch(left, right) => Branch(map(left)(f), map(right)(f))
  }
}

val tree: Tree[Int] = Branch(
  Branch(Leaf(1), Leaf(2)),
  Branch(Leaf(3), Leaf(4))
)
val stringTree = Functor[Tree].map(tree)(_.toString)
println(s"Tree functor: Int tree â†’ String tree")

// Lift functions to work on any functor
def addTen[F[_]: Functor](fa: F[Int]): F[Int] =
  Functor[F].map(fa)(_ + 10)

def toUpper[F[_]: Functor](fa: F[String]): F[String] =
  Functor[F].map(fa)(_.toUpperCase)

println(s"Add ten to list: ${addTen(List(1, 2, 3))}")
println(s"Add ten to option: ${addTen(Some(5))}")
println(s"Uppercase option: ${toUpper(Some("hello"))}")

println("\nâœ… Functor Laws:")
println("â€¢ Identity: fa.map(identity) == fa")
println("â€¢ Composition: fa.map(f.andThen(g)) == fa.map(f).map(g)")
println("â€¢ Functor doesn't change shape, just content")

println("\nðŸŽ¯ Functor Power:")
println("â€¢ Write once, work with List, Option, Either, Future, ...")
println("â€¢ Context-preserving transformation")
println("â€¢ Foundation for Monad and Applicative")
```

## ðŸ—ï¸ **Applicative: Independent Computations**

**Combine independent computations in any context**

```scala
// Applicative: Independent computation combination
import cats.Applicative

println("ðŸ—ï¸ APPLICATIVE: INDEPENDENT COMPUTATIONS IN CONTEXT")
println()

// Basic applicative usage
println("Basic Applicative Operations:")

// Pure - lift value into context
val listPure = Applicative[List].pure(42)
val optionPure = Applicative[Option].pure("hello")
println(s"List pure: $listPure")
println(s"Option pure: $optionPure")

// Applying functions in context
val add: (Int, Int) => Int = _ + _
val listAdd = Applicative[List].map2(List(1, 2), List(10, 20))(add)
val optionAdd = Applicative[Option].map2(Some(5), Some(3))(add)
println(s"List addition: $listAdd")
println(s"Option addition: $optionAdd")

// List of functions applied to list of values
val functions = List[Int => Int](_ + 1, _ * 2, _ - 3)
val values = List(10, 20, 30)
val applied = Applicative[List].ap(functions.map(Applicative[List].pure(_)))(values)
println(s"Function application: $applied")

// Form validation example
case class User(name: String, age: Int, email: String)

def validateName(name: String): Option[String] =
  if (name.length > 1) Some(name) else None

def validateAge(age: Int): Option[Int] =
  if (age >= 0 && age <= 150) Some(age) else None

def validateEmail(email: String): Option[String] =
  if (email.contains("@")) Some(email) else None

def createUser(name: Option[String], age: Option[Int], email: Option[String]): Option[User] =
  Applicative[Option].map3(name, age, email)(User.apply)

val goodUser = createUser(Some("Alice"), Some(25), Some("alice@email.com"))
val badUser = createUser(Some("A"), Some(-5), Some("bademail"))
println(s"Valid user: $goodUser")
println(s"Invalid user: $badUser")

// Applicative vs Monad comparison
println("\nðŸŽ¯ Applicative vs Monad:")
println("â€¢ Monad: Dependent computations (flatMap)")
println("â€¢ Applicative: Independent computations (map2, map3)")
println("â€¢ Use Applicative when computations don't depend on each other")
println("â€¢ All Monads are Applicatives (but not vice versa)")

println("\nâœ… Applicative Laws:")
println("â€¢ Identity: pure(id) <*> v == v")
println("â€¢ Homomorphism: pure(f) <*> pure(x) == pure(f(x)))")
println("â€¢ Interchange: u <*> pure(y) == pure(f => f(y)) <*> u")
println("â€¢ Composition: u <*> (v <*> w) == (u <*> v) <*> w")
```

## ðŸŒ€ **Monad: Sequential Computations**

**The most powerful abstraction - dependent computations with context**

```scala
// Monad: Sequential, dependent computations
import cats.Monad

println("ðŸŒ€ MONAD: SEQUENTIAL DEPENDENT COMPUTATIONS - THE CROWN JEWEL")
println()

// Basic monad operations
println("Basic Monad Operations:")

// Pure and flatMap
val monadList = Monad[List].pure(10)
val monadOption = Monad[Option].pure("hello")
println(s"List monad pure: $monadList")
println(s"Option monad pure: $monadOption")

// Conditional flatMap
def getUserOption(id: String): Option[String] =
  if (id == "123") Some("Alice") else None

def getEmailOption(name: String): Option[String] =
  if (name == "Alice") Some("alice@email.com") else None

// Using flatMap for dependent computations
val emailFromId = Monad[Option].flatMap(getUserOption("123")) { userName =>
  getEmailOption(userName)
}

val emailFromBadId = Monad[Option].flatMap(getUserOption("999")) { userName =>
  getEmailOption(userName)
}

println(s"Valid user email: $emailFromId")
println(s"Invalid user email: $emailFromBadId")

// For-comprehension (Scala sugar for flatMap)
def getUserEmail(id: String): Option[(String, String)] = {
  for {
    userName <- getUserOption(id)
    email <- getEmailOption(userName)
  } yield (userName, email)
}

println(s"For-comprehension result: ${getUserEmail("123")}")

// Identity monad example
val identityComputation = for {
  x <- Monad[List].pure(10)
  y <- List(x * 2, x + 5)
  z <- List(y, y * 10)
} yield z

println(s"Identity monad power: $identityComputation")

// Monad transformer for complex nesting
import cats.data.OptionT
import cats.effect.IO

// Function that returns IO[Option[String]]
def lookupUser(id: String): IO[Option[String]] =
  IO.pure(if (id == "123") Some("Alice") else None)

def lookupEmail(name: String): IO[Option[String]] =
  IO.pure(if (name == "Alice") Some("alice@email.com") else None)

// Using OptionT to flatten IO[Option[T]] -> IO[T]
def userEmailWorkflow(id: String): IO[Option[String]] = {
  (for {
    name <- OptionT(lookupUser(id))
    email <- OptionT(lookupEmail(name))
  } yield email).value
}

val workflowResult = userEmailWorkflow("123").unsafeRunSync()
println(s"Email workflow result: $workflowResult")

println("\nâœ… Monad Laws:")
println("â€¢ Left identity: pure(x).flatMap(f) == f(x)")
println("â€¢ Right identity: m.flatMap(pure) == m")
println("â€¢ Associativity: m.flatMap(f).flatMap(g) == m.flatMap(x => f(x).flatMap(g))")

println("\nðŸŽ¯ Monad Power:")
println("â€¢ Sequence dependent operations")
println("â€¢ Handle errors gracefully")
println("â€¢ Work with any context (IO, Future, Option, Either)")
println("â€¢ Composable side effects")
println("â€¢ Foundation of functional programming")
```

## ðŸ† **Cats Effect: Functional Effects**

**Pure functional programming for side effects**

```scala
// Cats Effect: Functional effect system
import cats.effect.{IO, Sync, Async, Timer, ContextShift}
import cats.effect.concurrent.Ref
import scala.concurrent.ExecutionContext
import scala.concurrent.duration._

println("ðŸ† CATS EFFECT: FUNCTIONAL SIDE EFFECTS")
println()

// Basic IO operations
val helloIO = IO(println("Hello, Cats Effect!"))
val delayedIO = IO.sleep(1.second) *> IO(println("After delay"))

println("Basic IO Operations:")
println("â€¢ IO.pure(x): Lift pure value into IO")
println("â€¢ IO.apply(x): Suspend side effect")
println("â€¢ IO.raiseError(e): Raise error")

// Functional error handling
def safeDivideIO(dividend: Int, divisor: Int): IO[Int] =
  if (divisor == 0)
    IO.raiseError(new ArithmeticException("Division by zero"))
  else
    IO(dividend / divisor)

val validDivision = safeDivideIO(10, 2)
val invalidDivision = safeDivideIO(10, 0)

println("Functional Error Handling:")
println("â€¢ handleErrorWith: Recover from errors")
println("â€¢ attempt: Convert Throwable to Either")
println("â€¢ redeern: Transform exception types")

// Resource management
import cats.effect.Resource
import java.io.{FileInputStream, File}

def readFileResource(file: File): Resource[IO, FileInputStream] =
  Resource.fromAutoCloseable(IO(new FileInputStream(file)))

val fileOperation = readFileResource(new File("example.txt")).use { inputStream =>
  IO(inputStream.read()) // Would handle file reading
}

println("\nResource Management:")
println("â€¢ Resource.make(acquire)(release): Bracket pattern")
println("â€¢ .use(f): Apply function and auto-cleanup")
println("â€¢ Safe acquisition and release")

// Concurrency primitives
def concurrentOperations(implicit cs: ContextShift[IO], timer: Timer[IO]): IO[Unit] = {
  val task1 = IO(println("Task 1 starting")) *> timer.sleep(500.millis) *> IO(println("Task 1 done"))
  val task2 = IO(println("Task 2 starting")) *> timer.sleep(300.millis) *> IO(println("Task 2 done"))

  for {
    fiber1 <- task1.start // Non-blocking
    fiber2 <- task2.start // Non-blocking
    _ <- fiber1.join     // Wait for completion
    _ <- fiber2.join     // Wait for completion
    result <- IO(println("All tasks completed"))
  } yield result
}

println("\nConcurrency with Fibers:")
println("â€¢ IO.start: Non-blocking fiber creation")
println("â€¢ Fiber.join: Wait for completion")
println("â€¢ Fiber.cancel: Safe cancellation")
println("â€¢ Race conditions eliminated by design")

// Mutable reference in pure functional style
def counterExample: IO[Unit] = {
  for {
    ref <- Ref.of[IO, Int](0)
    _ <- (0 to 10).toList.traverse_ { _ =>
      ref.update(_ + 1)
    }
    finalCount <- ref.get
    _ <- IO(println(s"Final count: $finalCount"))
  } yield ()
}

println("\nFunctional Mutability:")
println("â€¢ Ref: Mutable reference with atomic operations")
println("â€¢ Deferred: Single-assignment variable")
println("â€¢ Semaphore: Resource access control")
println("â€¢ All thread-safe and composable")

// IOApp for application entry point
println("\nðŸ”§ IOApp Pattern:")
println("object MyApp extends IOApp {")
println("  def run(args: List[String]): IO[ExitCode] = ...")
println("}")
println("\nBenefits:")
println("â€¢ Type-safe main method")
println("â€¢ Proper shutdown on interruption")
println("â€¢ CE3 compliance")
```

## ğŸš€ Problem 1: Linked List Cycle Detection (LeetCode #141)

### **ğŸ“‹ Problem Statement**

**ğŸ¯ Interview Question:**
*"Given head, the head of a linked list, determine if the linked list has a cycle in it."*

**ğŸ’¼ Business Context:**
Memory leak detection systems need to identify circular references, or cache invalidation systems detect infinite loops in data structures.

### **ğŸ“Š Input/Output**
```scala
Input: head = [3,2,0,-4], pos = 1  // pos indicates cycle start
Output: true  // Cycle exists

Input: head = [1,2], pos = -1  // no cycle
Output: false
```

### **ğŸ”’ Constraints**
- The number of nodes ranges from `0` to `10â´`
- `-10âµ â‰¤ Node.val â‰¤ 10âµ`
- `pos` is -1 or a valid index in the linked-list

### **ğŸ¯ Expected Approach**
Floyd's tortoise and hare algorithm with fast-slow pointers.

---

### **ğŸ’¡ Solution Approaches**

#### **Approach 1: Fast-Slow Pointers (Optimal)**
```scala
def hasCycle(head: Option[ListNode]): Boolean = {
  @tailrec
  def detect(slow: Option[ListNode], fast: Option[ListNode]): Boolean = {
    (slow, fast) match {
      case (Some(s), Some(f)) if s eq f => true  // Cycle detected
      case (Some(s), Some(f)) => f.next match {
        case Some(ff) => detect(s.next, ff.next)  // Fast moves 2x speed
        case None => false  // No cycle
      }
      case _ => false  // No cycle
    }
  }
  detect(head, head.flatMap(_.next))
}
```

#### **Approach 2: Hash Set (Extra Space)**
```scala
def hasCycleHashSet(head: Option[ListNode]): Boolean = {
  val visited = scala.collection.mutable.Set[ListNode]()
  
  @tailrec
  def traverse(current: Option[ListNode]): Boolean = {
    current match {
      case Some(node) if visited.contains(node) => true
      case Some(node) => 
        visited.add(node)
        traverse(node.next)
      case None => false
    }
  }
  traverse(head)
}
```

### **ğŸ¯ Interview Follow-ups**

1. **"What if you need to find the cycle start node?"**
   - After meeting point, reset slow to head, move both one step.

2. **"Can you solve without modifying the list?"**
   - Fast-slow pointers or hash set approaches.

3. **"What if there are multiple cycles?"**
   - Floyd's algorithm finds one cycle; problem assumes at most one.

4. **"How does the algorithm guarantee cycle detection?"**
   - Fast catches slow in O(Î¼ + Î») time where Î¼ is cycle start distance.

5. **"Real-world applications?"**
   - Garbage collection, deadlock detection, cache invalidation.

````scala
// Linked List Cycle Detection - Floyd's Algorithm
case class ListNode(value: Int, var next: Option[ListNode] = None)

def hasCycle(head: Option[ListNode]): Boolean = {
  @tailrec
  def detect(slow: Option[ListNode], fast: Option[ListNode]): Boolean = {
    (slow, fast) match {
      case (Some(s), Some(f)) if s eq f => true
      case (Some(s), Some(f)) => f.next.flatMap(_.next) match {
        case Some(ff) => detect(s.next, Some(ff))
        case None => false
      }
      case _ => false
    }
  }
  detect(head, head.flatMap(_.next))
}

// Helper to create list from array
def fromArray(arr: Array[Int]): Option[ListNode] = {
  arr.reverse.foldLeft(None: Option[ListNode]) { (tail, value) =>
    Some(ListNode(value, tail))
  }
}

// Test cases
println("=== Linked List Cycle Detection ===")
println("Floyd's tortoise and hare algorithm\n")

// Test 1: No cycle
val noCycle = fromArray(Array(1, 2, 3, 4, 5))
println(s"No cycle list: ${hasCycle(noCycle)}")

// Test 2: With cycle
val node1 = ListNode(3)
val node2 = ListNode(2)
val node3 = ListNode(0)
val node4 = ListNode(-4)
node1.next = Some(node2)
node2.next = Some(node3)
node3.next = Some(node4)
node4.next = Some(node2)  // Cycle
println(s"Cycle list: ${hasCycle(Some(node1))}")

println()
````

## ğŸš€ Problem 2: Reverse Linked List (LeetCode #206)

### **ğŸ“‹ Problem Statement**

**ğŸ¯ Interview Question:**
*"Given the head of a singly linked list, reverse the list, and return the reversed list."*

**ğŸ’¼ Business Context:**
Data processing pipelines need to reverse data streams, or undo operations require reversing changes in linked structures.

### **ğŸ”’ Constraints**
- The number of nodes in the list is the range `[0, 5000]`
- `-5000 â‰¤ Node.val â‰¤ 5000`

### **ğŸ¯ Expected Approach**
Iterative in-place reversal with three pointers.

---

### **ğŸ’¡ Solution Approaches**

#### **Approach 1: Iterative (Three Pointers)**
```scala
def reverseList(head: Option[ListNode]): Option[ListNode] = {
  var prev: Option[ListNode] = None
  var current = head
  
  while (current.isDefined) {
    val nextTemp = current.get.next
    current.get.next = prev
    prev = current
    current = nextTemp
  }
  
  prev
}
```

#### **Approach 2: Recursive**
```scala
def reverseListRecursive(head: Option[ListNode]): Option[ListNode] = {
  head match {
    case None => None
    case Some(node) if node.next.isEmpty => head
    case Some(node) =>
      val newHead = reverseListRecursive(node.next)
      node.next.get.next = Some(node)
      node.next = None
      newHead
  }
}
```

---

### **âš¡ Complexity Analysis**

| Approach | Time Complexity | Space Complexity | Notes |
|----------|----------------|------------------|-------|
| Iterative | O(n) | O(1) | **Optimal, in-place** |
| Recursive | O(n) | O(n) | Stack overflow risk |

### **ğŸ¯ Interview Follow-ups**

1. **"Can you reverse in groups of k?"**
   - Use sliding window approach, reverse each group.

2. **"What if you need to reverse doubly linked list?"**
   - Update both next and prev pointers.

3. **"How to reverse first k nodes?"**
   - Reverse first k, then attach rest.

4. **"Stack overflow concerns with recursion?"**
   - For n > 1000, use iterative approach.

5. **"Real-world applications?"**
   - Undo operations, LRU cache, browser back button.

---

````scala
// Reverse Linked List - Iterative Approach
def reverseList(head: Option[ListNode]): Option[ListNode] = {
  var prev: Option[ListNode] = None
  var current = head
  
  while (current.isDefined) {
    val nextTemp = current.get.next
    current.get.next = prev
    prev = current
    current = nextTemp
  }
  
  prev
}

// Test cases
println("=== Reverse Linked List ===")
println("Iterative in-place reversal\n")

val testCases = List(
  Array(1, 2, 3, 4, 5),
  Array(1, 2),
  Array(1),
  Array()
)

testCases.foreach { arr =>
  val original = fromArray(arr)
  val reversed = reverseList(original)
  
  val origStr = original.map(_.toList.mkString("[ ", " -> ", " ]")).getOrElse("[ ]")
  val revStr = reversed.map(_.toList.mkString("[ ", " -> ", " ]")).getOrElse("[ ]")
  
  println(s"$origStr â†’ $revStr")
}

println()
````

## ğŸš€ Problem 3: Merge Two Sorted Lists (LeetCode #21)

### **ğŸ“‹ Problem Statement**

**ğŸ¯ Interview Question:**
*"You are given the heads of two sorted linked lists list1 and list2. Merge the two lists in a one sorted list."*

**ğŸ’¼ Business Context:**
Merge sort algorithms combine sorted sublists, or data integration systems merge sorted datasets from multiple sources.

### **ğŸ”’ Constraints**
- The number of nodes in both lists is in the range `[0, 50]`
- `-100 â‰¤ Node.val â‰¤ 100`
- Both list1 and list2 are sorted in non-decreasing order

### **ğŸ¯ Expected Approach**
Iterative merge with dummy node for clean implementation.

---

````scala
// Merge Two Sorted Lists
def mergeTwoLists(l1: Option[ListNode], l2: Option[ListNode]): Option[ListNode] = {
  val dummy = ListNode(0)
  var current = dummy
  var list1 = l1
  var list2 = l2
  
  while (list1.isDefined && list2.isDefined) {
    if (list1.get.value <= list2.get.value) {
      current.next = list1
      current = current.next.get
      list1 = list1.get.next
    } else {
      current.next = list2
      current = current.next.get
      list2 = list2.get.next
    }
  }
  
  current.next = if (list1.isDefined) list1 else list2
  dummy.next
}

// Test cases
println("=== Merge Two Sorted Lists ===")
println("Using dummy node for clean iterative solution\n")

val mergeTests = List(
  (Array(1, 2, 4), Array(1, 3, 4)),
  (Array(), Array(0)),
  (Array(2), Array(1)),
  (Array(1, 3, 5), Array(2, 4))
)

mergeTests.foreach { case (arr1, arr2) =>
  val list1 = fromArray(arr1)
  val list2 = fromArray(arr2)
  
  val merged = mergeTwoLists(list1, list2)
  val result = merged.map(_.toList.mkString("[", " -> ", "]")).getOrElse("[]")
  
  println(s"${arr1.mkString("[", ", ", "]")} + ${arr2.mkString("[", ", ", "]")} â†’ $result")
}

println()
````

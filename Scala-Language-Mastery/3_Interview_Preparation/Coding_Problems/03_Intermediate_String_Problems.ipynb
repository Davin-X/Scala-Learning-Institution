### **ğŸ¯ Expected Approach**

### **ğŸ’¡ Solution Approaches**

#### **Approach 1: Functional (groupBy - Recommended)**
```scala
def findDuplicatesFunctional(str: String): Map[Char, Int] = {
  str.groupBy(identity).mapValues(_.length).filter(_._2 > 1)
}
```

#### **Approach 2: Mutable Map**
```scala
def findDuplicatesMutable(str: String): Map[Char, Int] = {
  val freq = scala.collection.mutable.Map[Char, Int]()
  str.foreach(c => freq(c) = freq.getOrElse(c, 0) + 1)
  freq.filter(_._2 > 1).toMap
}
```

#### **Approach 3: Array Frequency (ASCII only)**
```scala
def findDuplicatesArray(str: String): Map[Char, Int] = {
  val freq = Array.fill(256)(0)
  str.foreach(c => freq(c.toInt) += 1)
  (0 until 256).filter(i => freq(i) > 1).map(i => (i.toChar, freq(i))).toMap
}
```

## ğŸš€ Problem 2: Check if Strings are Anagrams

### **ğŸ“‹ Problem Statement**

**ğŸ¯ Interview Question:**
*"Given two strings s and t, return true if t is an anagram of s, and false otherwise."*

**ğŸ’¼ Business Context:**
Document similarity systems check if texts contain the same words, or security systems validate anagram-based passwords.

### **ğŸ”’ Constraints**
- `1 â‰¤ s.length, t.length â‰¤ 5 Ã— 10â´`
- `s` and `t` consist of lowercase English letters

### **ğŸ¯ Expected Approach**
Compare sorted strings or character frequency maps.

---

### **ğŸ’¡ Solution Approaches**

#### **Approach 1: Sort and Compare (Simple)**
```scala
def isAnagramSort(s: String, t: String): Boolean = {
  s.sorted == t.sorted
}
```

#### **Approach 2: Frequency Count (Optimal)**
```scala
def isAnagramFreq(s: String, t: String): Boolean = {
  if (s.length != t.length) return false
  
  val freq = Array.fill(26)(0)
  
  for (i <- s.indices) {
    freq(s(i) - 'a') += 1
    freq(t(i) - 'a') -= 1
  }
  
  freq.forall(_ == 0)
}
```

#### **Approach 3: Functional (groupBy)**
```scala
def isAnagramFunctional(s: String, t: String): Boolean = {
  s.groupBy(identity).mapValues(_.length) == t.groupBy(identity).mapValues(_.length)
}
```

---

### **âš¡ Complexity Analysis**

| Approach | Time Complexity | Space Complexity | Notes |
|----------|----------------|------------------|-------|
| Sort | O(n log n) | O(n) | Simple, works for any charset |
| Frequency | O(n) | O(1) | **Optimal for ASCII** |
| Functional | O(n) | O(min(n, charset)) | Clean, general purpose |

---

### **ğŸ¯ Interview Follow-ups**

1. **"What if strings contain Unicode characters?"**
   - Use Map[Char, Int] instead of array.

2. **"Case-sensitive or case-insensitive?"**
   - Clarify requirements; usually case-sensitive.

3. **"What if we need to find all anagram pairs in a list?"**
   - Group by sorted string or frequency signature.

4. **"Space optimization for very large strings?"**
   - Use streaming approach or external merge sort.

5. **"What if strings are very long but similar?"**
   - Early termination if lengths differ significantly.

---

````scala
// Check if two strings are anagrams
def isAnagram(str1: String, str2: String): Boolean = {
  str1.sorted == str2.sorted
}

def isAnagramFreq(str1: String, str2: String): Boolean = {
  if (str1.length != str2.length) return false
  
  val m1 = str1.groupBy(identity).mapValues(_.length)
  val m2 = str2.groupBy(identity).mapValues(_.length)
  
  m1.equals(m2)
}

// Test cases
println("=== Check if Strings are Anagrams ===")
println(s"word, wrdo: ${isAnagramFreq("word", "wrdo")}")
println(s"mary, army: ${isAnagram("mary", "army")}")
println(s"stop, tops: ${isAnagram("stop", "tops")}")
println(s"boat, btoa: ${isAnagram("boat", "btoa")}")
println(s"pure, in: ${isAnagram("pure", "in")}")  // Different lengths
println(s"fill, fil: ${isAnagramFreq("fill", "fil")}")
println()
````

## ğŸš€ Problem 3: First Non-Repeating Character

### **ğŸ“‹ Problem Statement**

**ğŸ¯ Interview Question:**
*"Given a string s, find the first non-repeating character in it and return its index. If it does not exist, return -1."*

**ğŸ’¼ Business Context:**
Text editors need to find unique characters for cursor placement, or data validation systems identify unique identifiers.

### **ğŸ”’ Constraints**
- `1 â‰¤ s.length â‰¤ 10âµ`
- `s` consists of only lowercase English letters

### **ğŸ¯ Expected Approach**
Two-pass algorithm: count frequencies, then find first with count 1.

---

### **ğŸ’¡ Solution Approaches**

#### **Approach 1: Two-Pass Frequency Count**
```scala
def firstUniqChar(s: String): Int = {
  val freq = Array.fill(26)(0)
  
  // Count frequencies
  s.foreach(c => freq(c - 'a') += 1)
  
  // Find first unique
  s.indices.find(i => freq(s(i) - 'a') == 1).getOrElse(-1)
}
```

#### **Approach 2: LinkedHashMap (Preserves Order)**
```scala
def firstUniqCharOrdered(s: String): Int = {
  val freq = scala.collection.mutable.LinkedHashMap[Char, Int]()
  
  // Count frequencies (preserves insertion order)
  s.foreach(c => freq(c) = freq.getOrElse(c, 0) + 1)
  
  // Find first with count 1
  freq.find(_._2 == 1).map { case (c, _) => s.indexOf(c) }.getOrElse(-1)
}
```

#### **Approach 3: Two Sets**
```scala
def firstUniqCharSets(s: String): Int = {
  val seen = scala.collection.mutable.Set[Char]()
  val duplicates = scala.collection.mutable.Set[Char]()
  
  s.foreach { c =>
    if (seen.contains(c)) {
      duplicates.add(c)
    } else {
      seen.add(c)
    }
  }
  
  s.indices.find(i => !duplicates.contains(s(i))).getOrElse(-1)
}
```

---

### **âš¡ Complexity Analysis**

| Approach | Time Complexity | Space Complexity | Notes |
|----------|----------------|------------------|-------|
| Two-Pass | O(n) | O(1) | **Optimal for ASCII** |
| LinkedHashMap | O(n) | O(min(n, 26)) | Preserves order |
| Two Sets | O(n) | O(min(n, 26)) | Simple logic |

### **ğŸ¯ Interview Follow-ups**

1. **"What if we need the character, not the index?"**
   - Return Option[Char] instead of Int.

2. **"What if there are multiple unique characters?"**
   - Problem asks for first one, but could return all.

3. **"How to handle Unicode characters?"**
   - Use Map[Char, Int] instead of array.

4. **"What if string is very large?"**
   - Two-pass is still optimal; streaming approach possible.

5. **"What if we need last non-repeating instead of first?"**
   - Reverse the search direction.

---

````scala
// Print first non-repeated character from String
def firstNonRepeating(input: String): Char = {
  val freq = scala.collection.mutable.Map[Char, Int]()
  
  // Count frequencies
  input.foreach(c => freq(c) = freq.getOrElse(c, 0) + 1)
  
  // Find first with count 1
  input.find(c => freq(c) == 1).getOrElse(' ')
}

// Test cases
println("=== First Non-Repeating Character ===")
println(s"Programming: ${firstNonRepeating("Programming")}")
println(s"CCommbination: ${firstNonRepeating("CCommbination")}")
println(s"aabbcc: ${firstNonRepeating("aabbcc")}")  // No unique chars
println()
````

## ğŸš€ Problem 4: Reverse String (Recursive)

### **ğŸ“‹ Problem Statement**

**ğŸ¯ Interview Question:**
*"Write a function that reverses a string. The input string is given as an array of characters s."*

**ğŸ’¼ Business Context:**
Text processing systems reverse strings for formatting, or data transformation pipelines reverse order for processing.

### **ğŸ”’ Constraints**
- `1 â‰¤ s.length â‰¤ 10âµ`
- `s[i]` is a printable ASCII character

### **ğŸ¯ Expected Approach**
Recursive function using head/tail decomposition.

---

### **ğŸ’¡ Solution Approaches**

#### **Approach 1: Recursive (Functional)**
```scala
def reverseString(str: String): String = {
  if (str.length <= 1) str
  else reverseString(str.tail) + str.head
}
```

#### **Approach 2: Built-in Method**
```scala
def reverseStringBuiltIn(str: String): String = {
  str.reverse
}
```

#### **Approach 3: Iterative (Two Pointers)**
```scala
def reverseStringIterative(chars: Array[Char]): Unit = {
  var left = 0
  var right = chars.length - 1
  
  while (left < right) {
    val temp = chars(left)
    chars(left) = chars(right)
    chars(right) = temp
    left += 1
    right -= 1
  }
}
```

---

### **âš¡ Complexity Analysis**

| Approach | Time Complexity | Space Complexity | Notes |
|----------|----------------|------------------|-------|
| Recursive | O(n) | O(n) | Functional, stack overflow risk |
| Built-in | O(n) | O(n) | Simple, creates new string |
| Iterative | O(n) | O(1) | **Optimal for arrays** |

### **ğŸ¯ Interview Follow-ups**

1. **"What if we can't modify the input?"**
   - Return new string instead of modifying array.

2. **"Reverse words instead of characters?"**
   - Split by space, reverse each word, join back.

3. **"What about Unicode characters?"**
   - Two-pointer approach works fine with any charset.

4. **"Stack overflow concerns with recursion?"**
   - For n > 10^5, use iterative approach.

5. **"What if we need to reverse in-place?"**
   - Two-pointer iterative is perfect for in-place.

---

````scala
// Reverse a given String using recursion
def reverseString(str: String): String = {
  if (str.length <= 1) str
  else reverseString(str.tail) + str.head
}

// Test cases
println("=== Reverse String (Recursive) ===")
println(s"Programming: ${reverseString("Programming")}")
println(s"Combination: ${reverseString("Combination")}")
println(s"Scala: ${reverseString("Scala")}")
println()
````

## ğŸš€ Problem 6: Count Vowels and Consonants

### **ğŸ“‹ Problem Statement**

**ğŸ¯ Interview Question:**
*"Write a program to count the number of vowels and consonants in a given string."*

**ğŸ’¼ Business Context:**
Text analysis systems categorize characters for language processing, or content filtering systems identify text characteristics.

### **ğŸ¯ Expected Approach**
Character classification and counting.

---

````scala
// Count vowels and consonants in a given String
def countVowelsConsonants(str: String): (Int, Int) = {
  val vowels = Set('a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U')
  
  val (vowelsCount, consonantsCount) = str.foldLeft((0, 0)) { case ((v, c), char) =>
    if (vowels.contains(char)) (v + 1, c)
    else if (char.isLetter) (v, c + 1)
    else (v, c)  // Ignore non-letters
  }
  
  (vowelsCount, consonantsCount)
}

// Test cases
println("=== Count Vowels and Consonants ===")
val (v1, c1) = countVowelsConsonants("Programming")
println(s"Programming: $v1 vowels, $c1 consonants")

val (v2, c2) = countVowelsConsonants("Combination")
println(s"Combination: $v2 vowels, $c2 consonants")

val (v3, c3) = countVowelsConsonants("Scala123!@#")
println(s"Scala123!@#: $v3 vowels, $c3 consonants")
println()
````

### **ğŸ“‹ Problem Statement**

**ğŸ’¼ Business Context:**

### **ğŸ¯ Expected Approach**
Simple counting with built-in methods or manual iteration.

---

````scala
// Count occurrence of a given character in String
def countOccurrences(str: String, elem: Char): Int = {
  str.count(_ == elem)
}

def countOccurrencesManual(str: String, elem: Char): Int = {
  var count = 0
  str.foreach(c => if (c == elem) count += 1)
  count
}

// Test cases
println("=== Count Character Occurrences ===")
println(s"Programming, 'g': ${countOccurrences("Programming", 'g')}")
println(s"CombinationCombination, 'i': ${countOccurrencesManual("CombinationCombination", 'i')}")
println(s"Scala, 'a': ${countOccurrences("Scala", 'a')}")
println()
````

## ğŸš€ Problem 10: Check String Rotation

### **ğŸ“‹ Problem Statement**

**ğŸ¯ Interview Question:**
*"Given two strings s and goal, return true if and only if s can become goal after some number of shifts on s."*

**ğŸ’¼ Business Context:**
Text comparison systems check if strings are rotations of each other, or circular buffer systems validate data arrangements.

### **ğŸ¯ Expected Approach**

````scala
// Check if two Strings are rotation of each other
def isRotation(str1: String, str2: String): Boolean = {
  str1.length == str2.length && (str1 + str1).contains(str2)
}

// Test cases
println("=== Check String Rotation ===")
println(s"IndiaVsAustralia, AustraliaVsIndia: ${isRotation("IndiaVsAustralia", "AustraliaVsIndia")}")
println(s"IndiaVsEngland, EnglandIndiaVs: ${isRotation("IndiaVsEngland", "EnglandIndiaVs")}")
println(s"hello, world: ${isRotation("hello", "world")}")
println()
````

## ğŸš€ Problem 11: Check Palindrome

### **ğŸ“‹ Problem Statement**

**ğŸ¯ Interview Question:**
*"Given a string s, return true if it is a palindrome, or false otherwise."*

**ğŸ’¼ Business Context:**
Text validation systems check palindrome properties, or data integrity systems validate symmetric structures.

### **ğŸ¯ Expected Approach**
Two-pointer comparison or built-in reverse.

---

````scala
// Check if the given String is Palindrome
def isPalindrome(str: String): Boolean = {
  str == str.reverse
}

// Two-pointer approach
def isPalindromeTwoPointer(str: String): Boolean = {
  val cleaned = str.toLowerCase.filter(_.isLetterOrDigit)
  var left = 0
  var right = cleaned.length - 1
  
  while (left < right) {
    if (cleaned(left) != cleaned(right)) return false
    left += 1
    right -= 1
  }
  true
}

// Test cases
println("=== Check Palindrome ===")
println(s"redivider: ${isPalindrome("redivider")}")
println(s"radar: ${isPalindromeTwoPointer("radar")}")
println(s"hello: ${isPalindrome("hello")}")
println(s"A man a plan a canal Panama: ${isPalindromeTwoPointer("A man a plan a canal Panama")}")
println()
````

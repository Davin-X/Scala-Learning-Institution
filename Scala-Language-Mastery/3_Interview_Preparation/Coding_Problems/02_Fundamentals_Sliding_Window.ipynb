## ğŸ—ï¸ Sliding Window Fundamentals

### **Core Concept**
Sliding window is an algorithmic technique that maintains a subset (window) of data and efficiently slides it through the entire dataset, often in O(n) time instead of O(nÂ²).

### **Key Patterns**
1. **Fixed Window**: Constant window size, track maximum/minimum/sum
2. **Variable Window**: Dynamic size based on conditions (expand/contract)
3. **Expanding Only**: Grow window until condition, then reset
4. **Deque Optimization**: Maintain monotonic properties for O(1) operations

### **Advantages**
- **Time Complexity**: Often O(n) vs O(nÂ²) brute force
- **Space Complexity**: O(1) to O(k) where k is window size
- **Versatility**: Solves 40% of array/string interview problems

### **ğŸ“‹ Problem Statement**

**ğŸ¯ Interview Question:**
*"Given an array of positive integers `nums` and a positive integer `target`, return the minimal length of a contiguous subarray `[numsl, numsl+1, ..., numsr-1, numsr]` of which the sum is greater than or equal to `target`. If there is no such subarray, return 0 instead."*

### **ğŸ”’ Constraints**
- `1 â‰¤ target â‰¤ 10â¹`
- `1 â‰¤ nums.length â‰¤ 10âµ`
- `1 â‰¤ nums[i] â‰¤ 10â´`

### **ğŸ¯ Expected Approach**

### **ğŸ’¡ Solution Approaches**

#### **Approach 1: Brute Force (O(nÂ³) time)**
```scala
def minSubarrayLenBruteForce(target: Int, nums: Array[Int]): Int = {
  val n = nums.length
  var minLength = Int.MaxValue
  
  // Check all possible subarrays
  for (i <- 0 until n) {
    for (j <- i until n) {
      val sum = nums.slice(i, j + 1).sum
      if (sum >= target) {
        minLength = math.min(minLength, j - i + 1)
      }
    }
  }
  
  if (minLength == Int.MaxValue) 0 else minLength
}
```

#### **Approach 2: Prefix Sum + Binary Search (O(n log n) time)**
```scala
def minSubarrayLenPrefixSum(target: Int, nums: Array[Int]): Int = {
  val prefix = nums.scanLeft(0)(_ + _).tail
  var minLength = Int.MaxValue
  
  for (i <- prefix.indices) {
    // Find smallest j such that prefix[j] - prefix[i-1] >= target
    val targetSum = prefix(i) + target
    val j = prefix.search(targetSum).insertionPoint
    if (j < prefix.length) {
      minLength = math.min(minLength, j - i + 1)
    }
  }
  
  if (minLength == Int.MaxValue) 0 else minLength
}
```

#### **Approach 3: Sliding Window (O(n) time)**
```scala
def minSubarrayLenSlidingWindow(target: Int, nums: Array[Int]): Int = {
  val n = nums.length
  var minLength = Int.MaxValue
  var currentSum = 0
  var left = 0
  
  for (right <- 0 until n) {
    currentSum += nums(right)
    
    // Contract window from left while sum >= target
    while (currentSum >= target && left <= right) {
      minLength = math.min(minLength, right - left + 1)
      currentSum -= nums(left)
      left += 1
    }
  }
  
  if (minLength == Int.MaxValue) 0 else minLength
}
```

````scala
// Minimum Size Subarray Sum - Variable sliding window
def minSubarrayLen(target: Int, nums: Array[Int]): (Int, Option[(Int, Int)]) = {
  val n = nums.length
  var minLength = Int.MaxValue
  var currentSum = 0
  var left = 0
  var bestWindow: Option[(Int, Int)] = None
  
  for (right <- 0 until n) {
    currentSum += nums(right)
    
    // Contract window from left while sum >= target
    while (currentSum >= target && left <= right) {
      val currentLength = right - left + 1
      if (currentLength < minLength) {
        minLength = currentLength
        bestWindow = Some((left, right))
      }
      currentSum -= nums(left)
      left += 1
    }
  }
  
  val resultLength = if (minLength == Int.MaxValue) 0 else minLength
  (resultLength, bestWindow)
}

// Test cases
val testCases = List(
  (7, Array(2, 3, 1, 2, 4, 3)),  // Expected: 2, Some((3,4)) -> [4,3]
  (4, Array(1, 4, 4)),           // Expected: 1, Some((1,1)) -> [4]
  (11, Array(1, 1, 1, 1, 1, 1, 1, 1)), // Expected: 0, None
  (3, Array(1, 2, 3))            // Expected: 1, Some((2,2)) -> [3]
)

println("=== Minimum Size Subarray Sum ===")
println("Variable sliding window: Expand right, contract left when sum >= target\n")

testCases.foreach { case (target, nums) =>
  val (length, window) = minSubarrayLen(target, nums)
  
  val windowElements = window.map { case (start, end) =>
    nums.slice(start, end + 1).mkString("[", ", ", "]")
  }.getOrElse("No solution")
  
  println(f"Target: $target%d, Array: ${nums.mkString("[", ", ", "]")}%s")
  println(f"Result: Length=$length%d, Window: $windowElements")
  
  // Verification
  window.foreach { case (start, end) =>
    val sum = nums.slice(start, end + 1).sum
    val valid = sum >= target
    println(f"Verification: Sum=${sum}%d >= target=${target}%d? $valid%b")
  }
  
  println("â”€" * 60)
}
println()
````

## ğŸš€ Problem 2: Longest Substring Without Repeating Characters (LeetCode #3)

### **ğŸ“‹ Problem Statement**

*"Given a string `s`, find the length of the longest substring without repeating characters."*

**ğŸ’¼ Business Context:**

### **ğŸ”’ Constraints**

### **ğŸ¯ Expected Approach**

### **ğŸ’¡ Solution Approaches**

#### **Approach 1: Brute Force (O(nÂ³) time)**
```scala
def lengthOfLongestSubstringBrute(s: String): Int = {
  var maxLength = 0
  
  for (i <- 0 until s.length) {
    for (j <- i until s.length) {
      val substring = s.substring(i, j + 1)
      if (substring.distinct.length == substring.length) {
        maxLength = math.max(maxLength, substring.length)
      }
    }
  }
  maxLength
}
```

#### **Approach 2: Sliding Window with HashSet (O(n) time)**
```scala
def lengthOfLongestSubstringHashSet(s: String): Int = {
  val charSet = collection.mutable.Set[Char]()
  var maxLength = 0
  var left = 0
  
  for (right <- s.indices) {
    while (charSet.contains(s(right))) {
      charSet.remove(s(left))
      left += 1
    }
    charSet.add(s(right))
    maxLength = math.max(maxLength, right - left + 1)
  }
  maxLength
}
```

#### **Approach 3: Sliding Window with ASCII Array (O(n) time, O(1) space)**
```scala
def lengthOfLongestSubstringASCII(s: String): Int = {
  val lastSeen = Array.fill(256)(-1)
  var maxLength = 0
  var left = 0
  
  for (right <- s.indices) {
    val charIndex = s(right).toInt
    if (lastSeen(charIndex) >= left) {
      left = lastSeen(charIndex) + 1
    }
    lastSeen(charIndex) = right
    maxLength = math.max(maxLength, right - left + 1)
  }
  maxLength
}
```

````scala
// Longest Substring Without Repeating Characters
def lengthOfLongestSubstring(s: String): (Int, String) = {
  val charSet = collection.mutable.Set[Char]()
  var maxLength = 0
  var maxSubstring = ""
  var left = 0
  
  for (right <- s.indices) {
    // Contract window until no duplicate
    while (charSet.contains(s(right))) {
      charSet.remove(s(left))
      left += 1
    }
    
    // Add current character
    charSet.add(s(right))
    
    // Update maximum
    val currentLength = right - left + 1
    val currentSubstring = s.substring(left, right + 1)
    
    if (currentLength > maxLength) {
      maxLength = currentLength
      maxSubstring = currentSubstring
    }
  }
  
  (maxLength, maxSubstring)
}

// Test cases
val substringTests = List(
  "abcabcbb",    // Expected: 3, "abc"
  "bbbbb",       // Expected: 1, "b"
  "pwwkew",      // Expected: 3, "wke"
  "",           // Expected: 0, ""
  "au",         // Expected: 2, "au"
  "abba"        // Expected: 2, "ab" or "ba"
)

println("=== Longest Substring Without Repeating Characters ===")
println("Sliding window with character set tracking\n")

substringTests.foreach { s =>
  val (length, substring) = lengthOfLongestSubstring(s)
  
  println(f"Input: \"$s%s\"")
  println(f"Result: Length=$length%d, Substring=\"$substring%s\"")
  
  // Verification
  val hasDuplicates = substring.distinct.length < substring.length
  val isValidLength = substring.length == length
  
  println(f"Verification: No duplicates=$hasDuplicates%b, Correct length=$isValidLength%b")
  
  if (hasDuplicates) {
    println("âŒ ERROR: Substring contains duplicate characters!")
  }
  
  println("â”€" * 50)
}
println()
````

## ğŸš€ Problem 3: Sliding Window Maximum (LeetCode #239)

### **ğŸ“‹ Problem Statement**

**ğŸ¯ Interview Question:**
*"You are given an array of integers `nums`, there is a sliding window of size `k` which is moving from the very left of the array to the very right. You can only see the `k` numbers in the window. Each time the sliding window moves right by one position. Return the max sliding window."*

**ğŸ’¼ Business Context:**
Financial systems need to track maximum stock prices in rolling time windows, or monitoring systems need to find peak values in sensor data streams.

### **ğŸ”’ Constraints**
- `1 â‰¤ nums.length â‰¤ 10âµ`
- `-10â´ â‰¤ nums[i] â‰¤ 10â´`
- `1 â‰¤ k â‰¤ nums.length`

### **ğŸ¯ Expected Approach**
Deque-based sliding window maximum for O(n) time.

---

### **ğŸ’¡ Solution Approaches**

#### **Approach 1: Brute Force (O(nÃ—k) time)**
```scala
def maxSlidingWindowBrute(nums: Array[Int], k: Int): Array[Int] = {
  val result = collection.mutable.ArrayBuffer[Int]()
  for (i <- 0 to nums.length - k) {
    val window = nums.slice(i, i + k)
    result.append(window.max)
  }
  result.toArray
}
```

#### **Approach 2: Deque (Monotonic Queue) - O(n) time**
```scala
def maxSlidingWindowDeque(nums: Array[Int], k: Int): Array[Int] = {
  val result = collection.mutable.ArrayBuffer[Int]()
  val deque = collection.mutable.ArrayDeque[Int]()
  
  for (i <- nums.indices) {
    // Remove elements outside current window
    while (deque.nonEmpty && deque.head < i - k + 1) {
      deque.removeHead()
    }
    
    // Maintain decreasing order
    while (deque.nonEmpty && nums(deque.last) <= nums(i)) {
      deque.removeLast()
    }
    
    deque.append(i)
    
    // Add max when window is complete
    if (i >= k - 1) {
      result.append(nums(deque.head))
    }
  }
  result.toArray
}
```

#### **Approach 3: Segment Tree (O(n log n) time)**
- Advanced data structure for range maximum queries
- Useful when multiple window sizes needed
- More complex but powerful for variations

---

### **âš¡ Complexity Analysis**

| Approach | Time Complexity | Space Complexity | When to Use |
|----------|----------------|------------------|-------------|
| Brute Force | O(nÃ—k) | O(1) | Small k, simple implementation |
| Deque | O(n) | O(k) | **Optimal for this problem** |
| Segment Tree | O(n log n) | O(n) | Multiple queries needed |

### **ğŸ¯ Interview Follow-ups**

1. **"What if we need minimum instead of maximum?"**
   - Same approach, just change comparison direction.

2. **"How does the deque optimization work?"**
   - Maintains indices in decreasing order for O(1) max access.

3. **"What if k > n?"**
   - Invalid input; return empty array or handle gracefully.

4. **"Can you do it with O(n) time and O(1) space?"**
   - No, impossible; need to track k elements somehow.

5. **"Real-world applications of this problem?"**
   - Time series analysis, signal processing, financial indicators.

````scala
// Sliding Window Maximum using Deque
def maxSlidingWindow(nums: Array[Int], k: Int): Array[Int] = {
  if (nums.isEmpty || k <= 0 || k > nums.length) return Array.emptyIntArray
  
  val result = collection.mutable.ArrayBuffer[Int]()
  val deque = collection.mutable.ArrayDeque[Int]()
  
  for (i <- nums.indices) {
    // Remove elements outside current window
    while (deque.nonEmpty && deque.head < i - k + 1) {
      deque.removeHead()
    }
    
    // Maintain decreasing order (remove smaller elements)
    while (deque.nonEmpty && nums(deque.last) <= nums(i)) {
      deque.removeLast()
    }
    
    // Add current element index
    deque.append(i)
    
    // Add maximum to result when window is complete
    if (i >= k - 1) {
      result.append(nums(deque.head))
    }
  }
  
  result.toArray
}

// Test cases
val slidingMaxTests = List(
  (Array(1, 3, -1, -3, 5, 3, 6, 7), 3),     // Expected: [3,3,5,5,6,7]
  (Array(1), 1),                               // Expected: [1]
  (Array(1, -1), 1),                           // Expected: [1,-1]
  (Array(9, 11), 2),                           // Expected: [11]
  (Array(4, -2), 2)                            // Expected: [4]
)

println("=== Sliding Window Maximum ===")
println("Using deque to maintain monotonic decreasing order\n")

slidingMaxTests.foreach { case (nums, k) =>
  val result = maxSlidingWindow(nums, k)
  
  println(f"Array: ${nums.mkString("[", ", ", "]")}%s, Window size: $k%d")
  println(f"Result: ${result.mkString("[", ", ", "]")}%s")
  
  // Verification
  val expected = (for (i <- 0 to nums.length - k) yield nums.slice(i, i + k).max).toArray
  val correct = result.sameElements(expected)
  
  println(f"Verification: Correct results = $correct%b")
  
  if (!correct) {
    println(f"Expected: ${expected.mkString("[", ", ", "]")}%s")
    println("âŒ ERROR: Results do not match!")
  }
  
  println("â”€" * 50)
}
println()
````

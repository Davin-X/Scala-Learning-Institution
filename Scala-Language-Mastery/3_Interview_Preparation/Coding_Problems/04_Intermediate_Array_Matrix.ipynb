## ğŸš€ Problem 1: Find Missing Number

### **ğŸ“‹ Problem Statement**

**ğŸ¯ Interview Question:**
*"Given an array nums containing n distinct numbers in the range [0, n], return the only number in the range that is missing from the array."*

**ğŸ’¼ Business Context:**
Database systems need to identify missing IDs in sequential ranges, or inventory systems detect missing product codes in catalogs.

### **ğŸ“Š Input/Output**
```scala
Input: nums = [3,0,1]
Output: 2  // n=3, missing number in [0,1,2,3]

Input: nums = [0,1]
Output: 2  // n=2, missing number in [0,1,2]
```

### **ğŸ”’ Constraints**
- `n == nums.length`
- `1 â‰¤ n â‰¤ 10â´`
- `0 â‰¤ nums[i] â‰¤ n`
- All numbers in nums are unique

### **ğŸ¯ Expected Approach**
Mathematical formula or XOR/bit manipulation.

---

### **ğŸ’¡ Solution Approaches**

#### **Approach 1: Mathematical Sum (O(n) time)**
```scala
def findMissingSum(nums: Array[Int]): Int = {
  val n = nums.length
  val expectedSum = n * (n + 1) / 2
  val actualSum = nums.sum
  expectedSum - actualSum
}
```

#### **Approach 2: XOR Method (O(n) time, O(1) space)**
```scala
def findMissingXOR(nums: Array[Int]): Int = {
  val n = nums.length
  var xor = 0
  
  // XOR all numbers from 0 to n
  for (i <- 0 to n) xor ^= i
  
  // XOR all numbers in array
  nums.foreach(xor ^= _)
  
  xor
}
```

#### **Approach 3: Hash Set (O(n) time, O(n) space)**
```scala
def findMissingSet(nums: Array[Int]): Int = {
  val n = nums.length
  val numSet = nums.toSet
  
  (0 to n).find(!numSet.contains(_)).get
}
```

---

### **âš¡ Complexity Analysis**

| Approach | Time Complexity | Space Complexity | Notes |
|----------|----------------|------------------|-------|
| Sum | O(n) | O(1) | Simple, may overflow for large n |
| XOR | O(n) | O(1) | **Optimal, handles overflow** |
| Hash Set | O(n) | O(n) | Straightforward, good for clarity |

---

## ğŸš€ Problem 2: Find Duplicate Numbers

### **ğŸ“‹ Problem Statement**

**ğŸ¯ Interview Question:**
*"Given an integer array nums, return all the duplicates that appear twice."*

**ğŸ’¼ Business Context:**
Data validation systems need to identify duplicate entries, or inventory systems detect double-counted items.

### **ğŸ”’ Constraints**
- `1 â‰¤ nums.length â‰¤ 10âµ`
- `1 â‰¤ nums[i] â‰¤ nums.length`

### **ğŸ¯ Expected Approach**
Frequency counting or in-place modification.

---

### **ğŸ’¡ Solution Approaches**

#### **Approach 1: Hash Map (O(n) time)**
```scala
def findDuplicatesHash(nums: Array[Int]): List[Int] = {
  val freq = scala.collection.mutable.Map[Int, Int]()
  nums.foreach(num => freq(num) = freq.getOrElse(num, 0) + 1)
  freq.filter(_._2 > 1).keys.toList.sorted
}
```

#### **Approach 2: Functional (groupBy)**
```scala
def findDuplicatesFunctional(nums: Array[Int]): List[Int] = {
  nums.groupBy(identity)
      .filter(_._2.length > 1)
      .keys.toList.sorted
}
```

#### **Approach 3: In-place Modification**
```scala
def findDuplicatesInPlace(nums: Array[Int]): List[Int] = {
  val result = scala.collection.mutable.ListBuffer[Int]()
  
  for (i <- nums.indices) {
    val index = math.abs(nums(i)) - 1
    if (nums(index) < 0) {
      result.append(math.abs(nums(i)))
    } else {
      nums(index) = -nums(index)
    }
  }
  
  result.toList
}
```

---

### **âš¡ Complexity Analysis**

| Approach | Time Complexity | Space Complexity | Notes |
|----------|----------------|------------------|-------|
| Hash Map | O(n) | O(n) | Simple and clear |
| Functional | O(n) | O(n) | Idiomatic Scala |
| In-place | O(n) | O(1) | **Space optimal** |

### **ğŸ¯ Interview Follow-ups**

1. **"What if you need to return duplicates with their counts?"**
   - Return Map[Int, Int] instead of List[Int].

2. **"Can you solve without modifying the array?"**
   - Use hash map or functional approach.

3. **"What if array contains negative numbers?"**
   - In-place approach won't work; use hash map.

4. **"How to find first duplicate only?"**
   - Stop after finding first element with count > 1.

5. **"Memory concerns for large arrays?"**
   - In-place approach is best for memory constraints.

---

````scala
// Find duplicate numbers in array
def findDuplicates(arr: Array[Int]): Map[Int, Int] = {
  arr.groupBy(identity).mapValues(_.length).filter(_._2 > 1)
}

// Test cases
println("=== Find Duplicate Numbers ===")
val result1 = findDuplicates(Array(1, 1, 2, 2, 3, 4, 5))
println(s"Array: [1,1,2,2,3,4,5] -> Duplicates: $result1")

val result2 = findDuplicates(Array(1, 1, 2, 2, 3, 4, 5))
println(s"Same array -> Duplicates: $result2")
println()
````

````scala
// Find largest and smallest number in unsorted array
def findMinMax(arr: Array[Int]): (Int, Int) = {
  val minVal = arr.min
  val maxVal = arr.max
  (minVal, maxVal)
}

// Alternative: Manual implementation
def findMinMaxManual(arr: Array[Int]): (Int, Int) = {
  var minVal = Int.MaxValue
  var maxVal = Int.MinValue
  
  arr.foreach { num =>
    if (num < minVal) minVal = num
    if (num > maxVal) maxVal = num
  }
  
  (minVal, maxVal)
}

// Test cases
println("=== Find Min/Max in Unsorted Array ===")
val testArray = Array(-20, 34, 21, -87, 92, Int.MaxValue, 5)
val (minVal, maxVal) = findMinMax(testArray)
println(s"Array: ${testArray.mkString("[", ", ", "]")}")
println(s"Min: $minVal, Max: $maxVal")

val (minManual, maxManual) = findMinMaxManual(testArray)
println(s"Manual verification - Min: $minManual, Max: $maxManual")
println()
````

## ğŸš€ Problem 4: Find Pair with Given Sum

### **ğŸ“‹ Problem Statement**

**ğŸ¯ Interview Question:**
*"Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target."*

**ğŸ’¼ Business Context:**
E-commerce systems find product combinations that meet price targets, or financial systems identify balancing transactions.

### **ğŸ”’ Constraints**
- `2 â‰¤ nums.length â‰¤ 10â´`
- `-10â¹ â‰¤ nums[i] â‰¤ 10â¹`
- `-10â¹ â‰¤ target â‰¤ 10â¹`
- Only one valid answer exists

### **ğŸ¯ Expected Approach**
Two pointers on sorted array or hash table.

---

````scala
// Find pair of integers whose sum equals given number
def findPair(arr: Array[Int], target: Int): Array[Int] = {
  val sorted = arr.sorted
  var left = 0
  var right = sorted.length - 1
  
  while (left < right) {
    val currentSum = sorted(left) + sorted(right)
    if (currentSum == target) {
      // Find original indices
      val idx1 = arr.indexOf(sorted(left))
      val idx2 = arr.lastIndexOf(sorted(right))
      return Array(idx1, idx2)
    } else if (currentSum < target) {
      left += 1
    } else {
      right -= 1
    }
  }
  Array.emptyIntArray
}

// Test cases
println("=== Find Pair with Given Sum ===")
println("Array: [7,9,8,10], Target: 15")
val result1 = findPair(Array(7, 9, 8, 10), 15)
println(s"Result: ${result1.mkString("[", ", ", "]")}")

println("Array: [3,4,7,5], Target: 9")
val result2 = findPair(Array(3, 4, 7, 5), 9)
println(s"Result: ${result2.mkString("[", ", ", "]")}")
println()
````

## ğŸš€ Problem 5: Find All Pairs with Given Sum

### **ğŸ“‹ Problem Statement**

**ğŸ¯ Interview Question:**
*"Given an array of integers and a target sum, find all unique pairs that add up to the target."*

**ğŸ’¼ Business Context:**
Recommendation systems find complementary products, or scheduling systems identify time slot combinations.

### **ğŸ¯ Expected Approach**
Hash set for O(n) time or sorting + two pointers.

---

````scala
// Find all pairs of integers whose sum equals given number
def findAllPairs(arr: Array[Int], target: Int): List[(Int, Int)] = {
  val seen = scala.collection.mutable.Set[Int]()
  val pairs = scala.collection.mutable.ListBuffer[(Int, Int)]()
  
  arr.foreach { num =>
    val complement = target - num
    if (seen.contains(complement)) {
      pairs.append((math.min(num, complement), math.max(num, complement)))
    }
    seen.add(num)
  }
  
  pairs.distinct.toList
}

// Test cases
println("=== Find All Pairs with Given Sum ===")
println("Array: [1,2,3,4,5,7,9,8,10], Target: 15")
val pairs1 = findAllPairs(Array(1, 2, 3, 4, 5, 7, 9, 8, 10), 15)
println(s"Pairs: $pairs1")

println("Array: [2,4,7,5,9,10,-1], Target: 9")
val pairs2 = findAllPairs(Array(2, 4, 7, 5, 9, 10, -1), 9)
println(s"Pairs: $pairs2")
println()
````

## ğŸš€ Problem 6: QuickSort Implementation

### **ğŸ“‹ Problem Statement**

**ğŸ¯ Interview Question:**
*"Implement the QuickSort algorithm to sort an integer array in place."*

**ğŸ’¼ Business Context:**
Database systems need efficient sorting algorithms, or analytics platforms require fast data ordering.

### **ğŸ¯ Expected Approach**
Recursive partitioning with pivot selection.

---

````scala
// QuickSort implementation
def quickSort(arr: Array[Int]): Unit = {
  def swap(i: Int, j: Int): Unit = {
    val temp = arr(i)
    arr(i) = arr(j)
    arr(j) = temp
  }
  
  def partition(low: Int, high: Int): Int = {
    val pivot = arr(high)
    var i = low - 1
    
    for (j <- low until high) {
      if (arr(j) <= pivot) {
        i += 1
        swap(i, j)
      }
    }
    swap(i + 1, high)
    i + 1
  }
  
  def sort(low: Int, high: Int): Unit = {
    if (low < high) {
      val pi = partition(low, high)
      sort(low, pi - 1)
      sort(pi + 1, high)
    }
  }
  
  sort(0, arr.length - 1)
}

// Functional QuickSort
def quickSortFunctional(arr: Array[Int]): Array[Int] = {
  if (arr.length <= 1) arr
  else {
    val pivot = arr(arr.length / 2)
    val (less, equal, greater) = (
      arr.filter(_ < pivot),
      arr.filter(_ == pivot),
      arr.filter(_ > pivot)
    )
    quickSortFunctional(less) ++ equal ++ quickSortFunctional(greater)
  }
}

// Test cases
println("=== QuickSort Implementation ===")
val arr1 = Array(1, 1, 2, 2, 3, 4, 5)
quickSort(arr1)
println(s"Array: ${arr1.mkString("[", ", ", "]")}")

val arr2 = Array(6, 5, 3, 1, 8, 7, 2, 4)
val sorted2 = quickSortFunctional(arr2)
println(s"Functional sort: ${sorted2.mkString("[", ", ", "]")}")
println()
````

## ğŸš€ Problem 7: Reverse Array

### **ğŸ“‹ Problem Statement**

**ğŸ¯ Interview Question:**
*"Write a function to reverse an array in place."*

**ğŸ’¼ Business Context:**
Data processing pipelines reverse data streams, or UI systems flip display orders for different layouts.

### **ğŸ¯ Expected Approach**
Two pointers swapping from ends.

---

````scala
// Reverse array in place
def reverseArray(arr: Array[Int]): Unit = {
  var left = 0
  var right = arr.length - 1
  
  while (left < right) {
    val temp = arr(left)
    arr(left) = arr(right)
    arr(right) = temp
    left += 1
    right -= 1
  }
}

// Recursive reverse
def reverseArrayRecursive(arr: Array[Int]): Array[Int] = {
  if (arr.length <= 1) arr
  else reverseArrayRecursive(arr.tail) :+ arr.head
}

// Test cases
println("=== Reverse Array ===")
val arr1 = Array(1, 1, 2, 2, 3, 4, 5)
reverseArray(arr1)
println(s"In-place: ${arr1.mkString("[", ", ", "]")}")

val arr2 = Array(6, 5, 3, 1, 8, 7, 2, 4)
val reversed2 = reverseArrayRecursive(arr2)
println(s"Recursive: ${reversed2.mkString("[", ", ", "]")}")
println()
````

## ğŸš€ Problem 8: Find Multiple Missing Numbers

### **ğŸ“‹ Problem Statement**

**ğŸ¯ Interview Question:**
*"Given an integer array with duplicates, find all numbers that are missing from the range 1 to max."*

**ğŸ’¼ Business Context:**
Inventory systems identify missing product IDs, or database systems find gaps in sequential keys.

### **ğŸ¯ Expected Approach**
Hash set comparison or sorting with iteration.

---

````scala
// Find multiple missing numbers in array
def findMultipleMissing(arr: Array[Int]): Array[Int] = {
  val maxVal = arr.max
  val present = arr.toSet
  val missing = scala.collection.mutable.ListBuffer[Int]()
  
  for (i <- 1 to maxVal) {
    if (!present.contains(i)) {
      missing.append(i)
    }
  }
  
  missing.toArray
}

// Test cases
println("=== Find Multiple Missing Numbers ===")
val missing1 = findMultipleMissing(Array(1, 1, 2, 3, 5, 5, 7, 9, 9, 9))
println(s"Array: [1,1,2,3,5,5,7,9,9,9] -> Missing: ${missing1.mkString("[", ", ", "]")}")

val missing2 = findMultipleMissing(Array(6, 5, 1, 8, 7, 2, 4))
println(s"Array: [6,5,1,8,7,2,4] -> Missing: ${missing2.mkString("[", ", ", "]")}")
println()
````

## ğŸš€ Problem 9: Binary Search

### **ğŸ“‹ Problem Statement**

**ğŸ¯ Interview Question:**
*"Implement binary search to find a target element in a sorted array."*

**ğŸ’¼ Business Context:**
Database indexes use binary search for fast lookups, or search engines find elements in sorted data structures.

### **ğŸ”’ Constraints**
- Array is sorted in ascending order
- `1 â‰¤ nums.length â‰¤ 10â´`
- `-10â´ â‰¤ nums[i], target â‰¤ 10â´`

### **ğŸ¯ Expected Approach**
Iterative or recursive binary search.

---

````scala
// Binary search implementation
def binarySearch(arr: Array[Int], target: Int): Boolean = {
  var left = 0
  var right = arr.length - 1
  
  while (left <= right) {
    val mid = left + (right - left) / 2
    
    if (arr(mid) == target) {
      return true
    } else if (arr(mid) < target) {
      left = mid + 1
    } else {
      right = mid - 1
    }
  }
  
  false
}

// Test cases
println("=== Binary Search ===")
println(s"Array: [1,1,2,2,3,4,5], Target: 2 -> ${binarySearch(Array(1,1,2,2,3,4,5), 2)}")
println(s"Array: [1,1,2,3,5,5,7,9,9,9], Target: 8 -> ${binarySearch(Array(1,1,2,3,5,5,7,9,9,9), 8)}")
println()
````

## ğŸš€ Problem 10: Matrix Transpose

### **ğŸ“‹ Problem Statement**

**ğŸ¯ Interview Question:**
*"Write a function to transpose a given matrix."*

**ğŸ’¼ Business Context:**
Data analysis systems transpose matrices for different views, or image processing systems rotate matrices for transformations.

### **ğŸ¯ Expected Approach**
Swap elements across diagonal or create new matrix.

---

````scala
// Matrix transpose
def transposeMatrix(matrix: Array[Array[Int]]): Array[Array[Int]] = {
  val rows = matrix.length
  val cols = matrix(0).length
  val transposed = Array.ofDim[Int](cols, rows)
  
  for (i <- 0 until rows; j <- 0 until cols) {
    transposed(j)(i) = matrix(i)(j)
  }
  
  transposed
}

// In-place transpose for square matrix
def transposeInPlace(matrix: Array[Array[Int]]): Unit = {
  val n = matrix.length
  for (i <- 0 until n; j <- i + 1 until n) {
    val temp = matrix(i)(j)
    matrix(i)(j) = matrix(j)(i)
    matrix(j)(i) = temp
  }
}

// Test cases
println("=== Matrix Transpose ===")
val matrix = Array(Array(1, 2, 3), Array(4, 5, 6))
val transposed = transposeMatrix(matrix)
println("Original matrix:")
matrix.foreach(row => println(row.mkString("[", ", ", "]")))
println("Transposed matrix:")
transposed.foreach(row => println(row.mkString("[", ", ", "]")))
println()
````

## ğŸš€ Problem 11: Matrix Addition

### **ğŸ“‹ Problem Statement**

**ğŸ¯ Interview Question:**
*"Write a function to add two matrices of the same dimensions."*

**ğŸ’¼ Business Context:**
Image processing systems combine multiple layers, or data analysis systems aggregate matrices.

### **ğŸ¯ Expected Approach**
Element-wise addition with dimension validation.

---

````scala
// Matrix addition
def addMatrices(m1: Array[Array[Int]], m2: Array[Array[Int]]): Array[Array[Int]] = {
  if (m1.length != m2.length || m1(0).length != m2(0).length) {
    throw new IllegalArgumentException("Matrices must have same dimensions")
  }
  
  val result = Array.ofDim[Int](m1.length, m1(0).length)
  for (i <- m1.indices; j <- m1(0).indices) {
    result(i)(j) = m1(i)(j) + m2(i)(j)
  }
  result
}

// Test cases
println("=== Matrix Addition ===")
val m1 = Array(Array(8, 9, 10), Array(14, 15, 16))
val m2 = Array(Array(10, 19, 20), Array(4, 5, 6))
val sum = addMatrices(m1, m2)
println("Matrix 1:")
m1.foreach(row => println(row.mkString("[", ", ", "]")))
println("Matrix 2:")
m2.foreach(row => println(row.mkString("[", ", ", "]")))
println("Sum:")
sum.foreach(row => println(row.mkString("[", ", ", "]")))
println()
````
